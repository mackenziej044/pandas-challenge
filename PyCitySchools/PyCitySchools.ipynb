## PyCity Schools Analysis

Analysis
_______________________________________________________________________________________________________________________________________
* The average reading scores at Charter and District schools fall around 82%, but only 81% of students pass reading in District Schools versus 97% in Charter Schools.

* Schools with larger student budgets underperformed compared to those with smaller budgets in every category. The more money spent per student, the lower the grades yielded. 
  
* Charter Schools performed exceptionally better than District Schools when calculating the Overall Pass Rate.
  
* Smaller Schools performed better than larger schools - this shows a correlation between class size and student scores.

```python
#import pandas
import pandas as pd
```

```python
#import data 
schooldatacsv = 'schools_complete.csv'
studentdatacsv = 'students_complete.csv'
```

```python
#read data 
schooldata_df = pd.read_csv(schooldatacsv)
studentdata_df = pd.read_csv(studentdatacsv)
```

```python
#merge into one dataset 
completedata_df = pd.merge(studentdata_df, schooldata_df, how='left', on=['school_name','school_name'])
completedata_df.head()
```

## District Summary

```python
#total number of schools 
schoolcount_df = schooldata_df['School ID'].unique()
schoolcount_df
```

```python
#convert array to numeric value
numberofschools = 15
```

```python
#total number of students
studentcount_df = schooldata_df['size'].sum()
studentcount_df
```

```python
#total budget 
budgettotal_df = schooldata_df['budget'].sum()
budgettotal_df
```

```python
#average math score
avgmath = completedata_df['math_score'].mean()
avgmath
```

```python
#average reading score
avgreading = completedata_df['reading_score'].mean()
avgreading
```

```python
#percentage of students who passed math (scores greater than or = 70)
passmath_count = completedata_df[(completedata_df['math_score']>=70)].count()['student_name']
passmath_percent = passmath_count/float(studentcount_df) * 100
passmath_percent
```

```python
#percentage of students who passed reading (scores greater than or = 70)
passreading_count = completedata_df[(completedata_df['reading_score']>=70)].count()['student_name']
passreading_percent = passreading_count/float(studentcount_df) * 100
passreading_percent
```

```python
#percentage of students who passed both reading and math
passboth_count = completedata_df[(completedata_df['math_score']>=70) & (completedata_df['reading_score']>=70)].count()['student_name']
overallpassrate = passboth_count/float(studentcount_df) * 100
overallpassrate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame({
    'Total Schools': [numberofschools],
    'Total Students':[studentcount_df],
    'Total Budget':[budgettotal_df],
    'Average Math Score':[avgmath],
    'Average Reading Score':[avgreading],
    '% Passing Math':[passmath_percent],
    '% Passing Reading':[passreading_percent],
    '% Overall Passing':[overallpassrate],
})

# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

# Display the DataFrame
district_summary
```

## School Summary

```python
#school count
schoolcount = len(completedata_df['school_name'].unique())
schoolcount
```

```python
#type per school from data
schooltype = schooldata_df.set_index(['school_name'])['type']
schooltype
```

```python
#total student per school
studentsperschool = completedata_df['school_name'].value_counts()
studentsperschool
```

```python
#total budget per school 
budgetperschool = completedata_df.groupby(['school_name'])['budget'].first()
budgetperschool
```

```python
#total budget per student (capita)
perschoolcapita = budgetperschool / schooldata_df.groupby('school_name')['size'].first()
perschoolcapita
```

```python
#average math scores per school 
avgmathps = completedata_df.groupby(['school_name'])['math_score'].mean()
avgmathps
```

```python
#average reading scores per school 
avgreadingps = completedata_df.groupby(['school_name'])['reading_score'].mean()
avgreadingps
```

```python
#number of students per school passing math
studentspassmath = completedata_df.loc[completedata_df['math_score']>=70]
studentspassmath

#math score by school 
totalmathscore = studentspassmath['school_name'].value_counts()

#percentage
percentmath = totalmathscore/studentsperschool*100
percentmath
```

```python
#number of students per school passing reading
studentspassread = completedata_df.loc[completedata_df['reading_score']>=70]
studentspassread

#reading score by school 
totalreadingscore = studentspassread['school_name'].value_counts()

#percentage
percentread = totalreadingscore/studentsperschool*100
percentread
```

```python
#calculate the number of students per school that passed both math and reading with scores of 70 or higher
studentspassboth = completedata_df[(completedata_df["reading_score"] >= 70) & (completedata_df["math_score"] >= 70)]
studentspassboth
schoolpassboth = studentspassboth.groupby(["school_name"]).size()
schoolpassboth
```

```python
#calculate passing rates 
perschoolmathpass = totalmathscore / schoolcount * 100

perschoolreadpass = totalreadingscore/schoolcount * 100

overallpass = schoolpassboth/studentsperschool * 100
```

```python
#create new DataFrame with columns from calculations above
per_school_summary = pd.DataFrame({
    'School Type':schooltype,
    'Total Students':studentsperschool,
    'Total School Budget':budgetperschool,
    'Per Student Budget':perschoolcapita,
    'Average Math Score':avgmathps,
    'Average Reading Score':avgreadingps,
    '% Passing Math':percentmath,
    '% Passing Reading':percentread,
    '% Overall Passing Rate':overallpass
})
# Formatting
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)

per_school_summary
```

## Highest Performing Schools (by % Overall Passing)

```python
#sort the schools in descending order / top 5 rows
top_schools = per_school_summary.sort_values(['% Passing Math', '% Passing Reading','% Overall Passing Rate'],ascending=False)
top_schools.head()
```

## Lowest Performing Schools (by % Overall Passing)

```python
#sort the schools in ascending order / bottom 5 rows
bottom_schools = per_school_summary.sort_values(['% Passing Math', '% Passing Reading','% Overall Passing Rate'],ascending=True)
bottom_schools.head()
```

## Scores By School Spending

```python
#establish bins
spending_bins = [0, 585, 630, 645, 680]
labels = ['<$585', '$585-630', '$630-645', '$645-680']
```

```python
#create a copy of the school summary - it has the "Per Student Budget"
spending_summary = per_school_summary.copy()
```

```python
# Convert 'Total School Budget' column to string type
per_school_summary['Total School Budget'] = per_school_summary['Total School Budget'].astype(str)

# Remove commas and dollar signs from the 'Total School Budget' column
per_school_summary['Total School Budget'] = per_school_summary['Total School Budget'].str.replace(',', '').str.replace('$', '')

# Convert 'Total School Budget' and 'Total Students' columns to numeric data types
per_school_summary['Total School Budget'] = pd.to_numeric(per_school_summary['Total School Budget'])
per_school_summary['Total Students'] = pd.to_numeric(per_school_summary['Total Students'])

# Bin the data based on school spending per student
spending_summary['Spending Ranges (Per Student)'] = pd.cut(per_school_summary['Total School Budget'] / per_school_summary['Total Students'], spending_bins, labels=labels)
spending_summary
```

```python
# Calculate the overall passing percentage based on passing rates for math and reading
spending_summary['% Overall Passing'] = (spending_summary['% Passing Math'] + spending_summary['% Passing Reading']) / 2

# Group by "Spending Ranges (Per Student)" and calculate the mean values
spending_math = spending_summary.groupby(["Spending Ranges (Per Student)"], observed=False)["Average Math Score"].mean()
spending_reading = spending_summary.groupby(["Spending Ranges (Per Student)"], observed=False)["Average Reading Score"].mean()
spending_passing_math = spending_summary.groupby(["Spending Ranges (Per Student)"], observed=False)["% Passing Math"].mean()
spending_passing_reading = spending_summary.groupby(["Spending Ranges (Per Student)"], observed=False)["% Passing Reading"].mean()
overall_passing_spending = spending_summary.groupby(["Spending Ranges (Per Student)"], observed=False)["% Overall Passing"].mean()
```

```python
#assemble into dataframe
# Create the spending_summary DataFrame
spending_summary = pd.DataFrame({
    "Average Math Score": spending_math,
    "Average Reading Score": spending_reading,
    "% Passing Math": spending_passing_math,
    "% Passing Reading": spending_passing_reading,
    "% Overall Passing": overall_passing_spending
})
spending_summary
```

## Scores by School Size

```python
#establish the bins
size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Categorize school sizes based on the total number of students
per_school_summary['School Size'] = pd.cut(per_school_summary['Total Students'], size_bins, labels=group_names)
per_school_summary
```

```python
# Calculate the overall passing percentage based on passing rates for math and reading
per_school_summary['% Overall Passing'] = (per_school_summary['% Passing Math'] + per_school_summary['% Passing Reading']) / 2

# Group by "School Size" and calculate the mean values
size_math = per_school_summary.groupby(["School Size"], observed=False)["Average Math Score"].mean()
size_reading = per_school_summary.groupby(["School Size"], observed=False)["Average Reading Score"].mean()
size_passing_math = per_school_summary.groupby(["School Size"], observed=False)["% Passing Math"].mean()
size_passing_reading = per_school_summary.groupby(["School Size"], observed=False)["% Passing Reading"].mean()
size_overall_passing = per_school_summary.groupby(["School Size"], observed=False)["% Overall Passing"].mean()
```

```python
#create dataframe
# Create the size_summary DataFrame
size_summary = pd.DataFrame({
    "Average Math Score": size_math,
    "Average Reading Score": size_reading,
    "% Passing Math": size_passing_math,
    "% Passing Reading": size_passing_reading,
    "% Overall Passing": size_overall_passing
})
size_summary
```
